## 1. As propriedades óticas de nuvens individuais dependem do Caminho de Água Líquida de Nuvem
____________


Considere $w$ como uma representação do conteúdo de água líquida em uma unidade de volume de ar de uma nuvem, em unidade de g m$^{-3}$.

O Caminho de Água Líquida (Liquid Water Path) da nuven é

$$LWP = w ~ \Delta z$$

onde $\Delta z$ é a profundidade da camada de nuvem em metros.  $LWP$ possui unidades de g m$^{-2}$.

$LWP$ determina a principal propriedade ótica da nuvem, tanto em termos de radiação de ondas longas e de ondas curtas:

<img src='../images/CloudOpticalProperties_Webster1994.png'>

Pontos importantes sobre as propriedades óticas da água de nuvens:

- emissividade de onda longa / absorvidade aumenta rapidamente com $LWP$
- o albedo da nuvem aumenta devagar com $LWP$

### Efeitos da radiação de onda longa e as nuvens

Como a emissividade satura para o caso de nuvens não muito espessas, nuvens de maior espessura comportam-se como um absorvedor de corpo negro em todos os níveis. Emissões a partir de baixo e dentro da nuvem será absorvida pela camada de cima da nuvem.

Emissões para o espaço são portanto **governadas pelo topo da nuvem**.

Os efeitos de radiação de onda longa de uma nuvem espessa depende **da temperatura do topo da nuvem**. Esta temperatura é determinada principalmente pela **altura do topo da nuvem**.

Uma nuvem com **topo alto** exercerá um efeito estufa forte porque ela absorve radiação que flui para cima e re-emite esta radiação em sua temperatura fria.

A radiação de onda longa destas nuvens tende a **aquecer** a superfície.

### Efeitos da radiação de onda curta e as nuvens

Como as nuvens aumentam o albedo planetário, os efeitos das nuvens na radiação de onda curta tende a resfriar a superfície.

A mesma nuvem portanto induz o balanço de energia planetária em duas direções simultaneamente. Qual efeito domina depende:

- da temperatura do topo da nuvem relativa á superfície
- do caminho de água líquida da nuvem LWP (profundidade da nuvem)

**Nuvens finas** são relativamente transparentes á radiação solar. Nuvens mais **espessas** são refletores efetivos.

Uma nuvem **cirrus**, por exemplo, possui um albedo negligível mas exerce um papel substancial no efeito estufa porque localiza-se próximo da fria tropopausa.Estas nuvens possuem um efeito resultante de aquecimento.

Uma nuvem **stratus** espessa localizada no topo da camada limite planetária reflete grandes quantidades de radiação solar incidente. No entanto, a temperatura no topo da nuvem não é muito diferente da temperatura da superfície, de forma que o efeito estufa é desprezível(mesmo que a nuvem seja forte absorvedor de radiação de onda longa!)

## 2. Efeitos na radiação da cobertura de nuvens versus céu claro
____________

Seja $F = ASR - OLR$ o balanço de radiação no topo da atmosfera (TOA).

Suponha que o fluxo médio na porção de céu claro **sem nuvens** seja $F_{clear}$.

Denominaremos o fluxo na região com nuvens por $F_{cloudy}$.

O fluxo total é a soma

$$ F = (1-c) F_{clear} + c F_{cloudy} $$

onde $0 \le c \le 1$ é a **fração de nuvem**, i.e. a fração do céu coberto pela nuvem.

Podemos quebrar esta estimativa nas componentes de radiação de onda longa $F_{LW}$ e onda curta $F_{SW}$:

$$ F = F_{LW} + F_{SW} $$

$$ F_{LW} = - \big((1-c)~OLR_{clear} + c ~ OLR_{cloudy} \big)$$

$$ F_{SW} = + \big((1-c)~ASR_{clear} + c ~ ASR_{cloudy} \big)$$

Esta nuvem vai aquecer a superfície se $F_{cloudy} > F_{clear}$, neste caso o fluxo resultante $F$ vai aumentar com o aumento da fração de nuvens $c$.

No exemplo acima, podemos admitir o seguinte:

#### Nuvem cirrus alta
- $ASR_{cloudy} \approx ASR_{clear} $
- $OLR_{cloudy} < OLR_{clear}$
- $F$ aumenta com $c$ (estas nuvens aquecem a superfície)

#### Nuvem stratus alta
- $ASR_{cloudy} < ASR_{clear} $
- $OLR_{cloudy} \approx OLR_{clear}$
- $F$ diminui com $c$ (estas nuvens resfriam a superfície)

Muitas outras nuvens possuem comportamento ambíguo. Por exemplo:

#### Nuvem cumulonimbus

- $ASR_{cloudy} < ASR_{clear} $
- $OLR_{cloudy} < OLR_{clear}$
- $F$ pode ou aumentar ou diminuir com $c$

Precisamos de um modelo para trabalhar estes detalhes!

## 3. Efeitos Radiativos da Nuvem (CRE)
____________

Geralmente não temos apenas um tipo de nuvem, mas muitos tipos que devemos considerar simultaneamente, seja nas observações da natureza(e.g.satelites) ou em um modelo GCM.

Na prática, raramente calculamos $F_{cloudy}$ explicitamente.

Ao invés disto definimos o **Efeito Radiativo da Nuvem** como

$$ CRE =  F - F_{clear} $$

que pode ser descrito em termos da fração de nuvem:

$$ CRE = c \big( F_{cloudy} - F_{clear} \big) $$

No exemplo acima, $CRE$ é positivo para cirrus, negativo para stratus baixos, e desconhecido para cumulonimbus.

Calculamos CRE (incluindo ambas as componentes de radiação de longa longa e onda curta) nas simulações do modelo.

## 4. Modelagem da dependencia do CRE em relação ao topo da nuvem

Vamos usar o modelo de radiação `RRTMG` para calcular o efeito radiativo em um modelo de uma coluna simples, e olhar como o CRE depende das propriedades da nuvem e da altura da cobertura da camada nuvem.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import climlab
from climlab.radiation import RRTMG
```

### Temperatura e umidade específica global média observadas

```python
#  Get temperature and humidity data from NCEP Reanalysis
import xarray as xr
ncep_url = "http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/ncep.reanalysis.derived/pressure/"
path = ncep_url
ncep_air = xr.open_dataset(path + 'air.mon.1981-2010.ltm.nc', decode_times=False)
ncep_shum = xr.open_dataset(path + 'shum.mon.1981-2010.ltm.nc', decode_times=False)
## The NOAA ESRL server is shutdown! January 2019
#url = 'http://apdrc.soest.hawaii.edu:80/dods/public_data/Reanalysis_Data/NCEP/NCEP/clima/pressure/'
#air = xr.open_dataset(url + 'air')
#shum = xr.open_dataset(url + 'shum')
#ncep_air = air.rename({'lev': 'level'})
#ncep_shum = shum.rename({'lev': 'level'})
```

```python
#  Take global, annual average and convert to correct units (Kelvin and kg/kg)
weight = np.cos(np.deg2rad(ncep_air.lat)) / np.cos(np.deg2rad(ncep_air.lat)).mean(dim='lat')
Tglobal = (ncep_air.air * weight).mean(dim=('lat','lon','time')) + climlab.constants.tempCtoK
SHglobal = (ncep_shum.shum * weight).mean(dim=('lat','lon','time')) * 1E-3  # kg/kg
```

Para estabelecer o modelo de radiação com diferentes camadas de pressão, teremos que fazer algumas interpolações verticais.

```python
#  Create a state dictionary with 50 levels
state = climlab.column_state(num_lev=50)
lev = state.Tatm.domain.axes['lev'].points
```

```python
# interpolate to model pressure levels
Tinterp = np.interp(lev, np.flipud(Tglobal.level), np.flipud(Tglobal))
SHinterp = np.interp(lev, np.flipud(SHglobal.level), np.flipud(SHglobal))
#  Need to 'flipud' because the interpolation routine
#  needs the pressure data to be in increasing order
```

```python
#  Plot the temperature and humidity profiles
fig, ax1 = plt.subplots(figsize=(8,5))
Tcolor = 'r'
SHcolor = 'b'
ax1.plot(Tinterp, lev, color=Tcolor)
ax1.invert_yaxis()
ax1.set_xlabel('Temperature (K)', color=Tcolor)
ax1.tick_params('x', colors=Tcolor)
ax1.grid()
ax1.set_ylabel('Pressure (hPa)')
ax2 = ax1.twiny()
ax2.plot(SHinterp*1E3, lev, color=SHcolor)
ax2.set_xlabel('Specific Humidity (g/kg)', color=SHcolor)
ax2.tick_params('x', colors=SHcolor)
fig.suptitle('Global mean air temperature and specific humidity', y=1.03, fontsize=14)
```

```python
#  Set the temperature to the observed values
state.Tatm[:] = Tinterp
```

```python
#  Define some local cloud characteristics
#  We are going to repeat the calculation
#   for three different types of clouds:
#   thin, medium, and thick
cldfrac = 0.5  # layer cloud fraction
r_liq = 14.  # Cloud water drop effective radius (microns)
# in-cloud liquid water path (g/m2)
clwp = {'thin': 20.,
        'med': 60.,
        'thick': 200.,}
```

```python
#  Loop through three types of cloud
#  for each type, loop through all pressure levels
#  Set up a radiation model with the cloud layer at the current pressure level
#  Compute CRE and store the results
CRE_LW = {}
CRE_SW = {}
for thickness in clwp:
    OLR = np.zeros_like(lev)
    ASR = np.zeros_like(lev)
    OLRclr = np.zeros_like(lev)
    ASRclr = np.zeros_like(lev)
    for i in range(lev.size):
        # Whole-column cloud characteristics
        #  The cloud fraction is a Gaussian bump centered at the current level
        mycloud = {'cldfrac': cldfrac*np.exp(-(lev-lev[i])**2/(2*25.)**2),
                   'clwp': np.zeros_like(state.Tatm) + clwp[thickness],
                   'r_liq': np.zeros_like(state.Tatm) + r_liq,}
        rad = RRTMG(state=state,
                    albedo=0.2,
                    specific_humidity=SHinterp,
                    verbose=False,
                    **mycloud)
#        rad.compute_diagnostics()
        OLR[i] = rad.OLR
        OLRclr[i] = rad.OLRclr
        ASR[i] = rad.ASR
        ASRclr[i] = rad.ASRclr
    CRE_LW[thickness] = -(OLR - OLRclr)
    CRE_SW[thickness] = (ASR - ASRclr)
```

```python
#  Make some plots of the CRE dependence on cloud height
fig, axes = plt.subplots(1,3, figsize=(16,6))
ax = axes[0]
for thickness in clwp:
    ax.plot(CRE_LW[thickness], lev, label=thickness)
ax.set_ylabel('Pressure (hPa)')
ax.set_xlabel('LW cloud radiative effect (W/m2)')

ax = axes[1]
for thickness in clwp:
    ax.plot(CRE_SW[thickness], lev, label=thickness)
ax.set_xlabel('SW cloud radiative effect (W/m2)')

ax = axes[2]
for thickness in clwp:
    ax.plot(CRE_SW[thickness] + CRE_LW[thickness], lev, label=thickness)
ax.set_xlabel('Net cloud radiative effect (W/m2)')

for ax in axes:
    ax.invert_yaxis()
    ax.legend()
    ax.grid()
fig.suptitle('Cloud Radiative Effect as a function of the vertical height of the cloud layer', fontsize=16)
```

Oque podemos concluir a partir destes resultados? Verifique como os efeitos das nuvens afeta a Radiação de Onda Longa e Onda Curta em função das propriedades e altura das nuvens.

## 5. Retroalimentação (Feedback) das Nuvens
____________

$CRE$ (o efeito radiativo das nuvens) depende de duas propriedades:

- a fração de nuvem $c$
- o caminho de água líquida (LWP), que determina $F_{cloudy}$

Se algum destes parâmetros (ou ambos) mudar conforme ocorrer uma mudança climática e houver um aquecimento da superfície, então existe uma fonte adicional de energia no topo da atmosfera TOA que ajuda a determinar o equilíbrio final de aquecimento --  um mecanismo de retroalimentação (feedback)!

O efeito de retroalimentação das nuvens (feedback) depende portanto, de mudanças na frequencia de ocorrência e das propriedades óticas de cada um dos diferentes tipos de nuvem. Este é um problema de grande complexidade.

Seguindo a notação de sensitividade e retroalimentação vistos anteriormente [veja](https://brian-rose.github.io/ClimateLaboratoryBook/courseware/sensitivity-feedback.html), podemos escrever a mudança no balanço de energia no topo da atmosfera após a resposta a uma forçante radiativa $\Delta R$ como

$$ \Delta F = \Delta R - \Delta T \left( \lambda_0 - \sum_{i=1}^N \lambda_i \right)  $$

onde $\lambda_0$ é a "retroalimentação de Planck (feedback)", ou, mais precisamente, o parâmetro de resposta de não-retroalimentação, enquanto o $\lambda_i$ são todos os parâmetros aditivos de retro-alimentação.

Agora usando

$$ F = (1-c) F_{clear} + c F_{cloudy} $$

podemos separar a mudança de $F$ em componentes devido á mudanças da fração de nuvens, fluxos em céu claro, e propriedades óticas:

$$ \Delta F = (1-c)\Delta F_{clear}  + c \Delta F_{cloudy}  +\left( F_{cloudy} - F_{clear} \right) \Delta c $$

onde $c, F_{cloudy}, F_{clear}$ aqui é estimado a partir do valor do clima de referência (controle), e assumimos que as mudanças são pequenas de forma que a linearização é sensível.

### Céu clado e a retroalimentação (feedbacks) das nuvens

A estimativa para a variação no TOA $\Delta F$ fornece

$$ (1-c)\Delta F_{clear}  + c \Delta F_{cloudy}  +\left( F_{cloudy} - F_{clear} \right) \Delta c = \Delta R - \Delta T \left( \lambda_0 - \sum_{i=1}^N \lambda_i \right) $$


We can break down the net feedback in many different ways, but here we will just conceptually separate clear-sky and cloud feedbacks:

$$ \sum_{i=1}^N \lambda_i = \lambda_{clear} + \lambda_{cloud} $$

A sensible definition of $\lambda_{cloud}$ would be involve **only changes in cloud fraction and properties**. We thus formally define

$$ \lambda_{cloud} = c \frac{\Delta F_{cloudy}}{\Delta T}  +\left( F_{cloudy} - F_{clear} \right) \frac{\Delta c}{\Delta T} $$

which we subtract from the net top-of-atmosphere budget to get

$$ (1-c)\Delta F_{clear}  = \Delta R - \Delta T \left( \lambda_0 - \lambda_{clear} \right) $$



which suggests that we can estimate the **clear-sky feedback** from

$$ \lambda_{clear} = \lambda_0 - \frac{\Delta R}{\Delta T} + (1-c) \frac{\Delta F_{clear}}{\Delta T}$$

This includes processes such as lapse rate feedback, water vapor feedback, and surface albedo feedback.

Remember that all of these expressions can be (and frequently are) decomposed into longwave and shortwave components.

### Retroalimentação de nuvens (feedback) vs. CRE (Efeito radiativo das nuvens)

Diagnósticos de modelos GCM geralmente produzem $CRE$ (que são computados por simulações com e sem a fração de nuvens).

Podemos calcular a **mudança no $CRE$** entre uma simulação climática de controle e outra perturbada.

Note que neste caso **a mudança em $CRE$ não é equivalente a uma retroalimentação**.

Note que
$$ \Delta CRE = \Delta F - \Delta F_{clear} $$

Com esta definição, podemos escrever como

$$ \Delta CRE = -c \Delta F_{clear}  + \Delta T \lambda_{cloud} $$

A retroalimentação de **céu claro** afeta a mudança em $CRE$ e podemos avaliar em GCM ou observações.
