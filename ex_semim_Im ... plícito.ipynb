```python
import numpy as np
import matplotlib.pyplot as plt 
import math

C1 = 100. #solucao explícita
C2 = 100. #solucao implicida
C3 = 100. #solucao exata ou analítica
C4 = 100. #solucao semi implícita

################################################################################

# parametro físico (arbitrario) 
lamb = 1.e-4

################################################################################

# parametros do modelo 
dt = 0.1/lamb
ntot = 40
time = 0.0
efold = 1.0/lamb 
E= 0.08 #com o E(epsilon) muito grande o C4 extrapola
print('time ', 'C1','C2','C3','C4')
for it in range ( 0, ntot + 1 ):
  time = it*dt # avanço no tempo 
  print(time/efold,C1,C2,C3,C4)
  C1N = C1*(1.0 - dt*lamb) # esquema explicito 
  C2N = C2/(1.0 + dt*lamb) # esquema implicito
  C4N = (C4-((1 - E)*dt*lamb*C4))/1 + E*dt*lamb #híbrido

################################################################################

  # Avança o passo de tempo
  C1 = C1N
  C2 = C2N
  C4 = C4N

################################################################################

  # solução analitica 
  xx=lamb*time
  C3N = 100.*math.exp(-xx)
  C3=C3N
```
